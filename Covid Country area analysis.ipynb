```python
from sklearn.preprocessing import LabelEncoder
l_en = LabelEncoder()
covid['Deaths / 100 Cases'] = l_en.fit_transform(covid['Deaths / 100 Cases'])
print(covid['Deaths / 100 Cases'])
```

```python
from sklearn.preprocessing import LabelEncoder
l_en = LabelEncoder()
covid['Recovered / 100 Cases'] = l_en.fit_transform(covid['Recovered / 100 Cases'])
print(covid['Recovered / 100 Cases'])
```

```python
from sklearn.preprocessing import LabelEncoder
l_en = LabelEncoder()
covid['Deaths / 100 Recovered'] = l_en.fit_transform(covid['Deaths / 100 Recovered'])
print(covid['Deaths / 100 Recovered'])
```

```python
X= covid[['Confirmed', 'Deaths', 'Recovered', 'Active',
       'New cases', 'New deaths', 'New recovered', 'Deaths / 100 Cases',
       'Recovered / 100 Cases', 'Deaths / 100 Recovered',
       'Confirmed last week', '1 week change', '1 week % increase']]
Y= covid[['Deaths / 100 Cases', 'Recovered / 100 Cases', 'Deaths / 100 Recovered','1 week % increase']]
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 300)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
model2.fit(X_train, Y_train)
```

```python
y_test_predicted =model2.predict(X_test)
print(y_test_predicted)
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_test, y_test_predicted)*100)
```
